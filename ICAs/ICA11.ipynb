````python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

### Static Plot, Mass vs. Distance

````python
data_nona = data.dropna(subset=["MASS", "DIST"])
data_nona = data_nona.iloc[1:,:]
````

````python
data_nona = data_nona.astype({
    "MASS": "float64",
    "DIST": "float64"
})
````

````python
plt.scatter(data_nona["MASS"], data_nona["DIST"], alpha=0.4)
plt.xlabel("Planet Mass (unit unknown)")
plt.ylabel("Planet Distance (unit unknown)")
plt.title("Exoplanet Mass vs. Distance")
plt.show()
````

### Attempt at Animating, Orbital Period of planet vs. Mass of Star (revealed in increments of time of periastron)

````python
data_nona_OPSMT = data.dropna(subset=["T0", "PER", "MSTAR"])
data_nona_OPSMT = data_nona_OPSMT.iloc[1:,:]
data_nona_OPSMT = data_nona_RM.astype({
    "T0": "float64",
    "MSTAR": "float64",
    "PER": "float64"
})
````

````python
plt.scatter(data_nona_OPSMT["MSTAR"], data_nona_OPSMT["PER"], alpha=0.4)
plt.xlabel("Mass of Star")
plt.ylabel("Orbital Period")
plt.title("Mass of Star vs. Orbital Period")
plt.yscale("log")
````

````python
t0_min = data_nona_OPSMT["T0"].min()
t0_max = 2458468.424617899 #hard coded, but found based on trial and error
t0_step = np.linspace(t0_min, t0_max, 100)

fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(1,1,1)

for i, t0 in enumerate(t0_step):

    curr_data = data_nona_OPSMT[data_nona_OPSMT["T0"] > t0]

    ax.scatter(curr_data["MSTAR"], curr_data["PER"], alpha=0.4)
    ax.set_xlim(-0.25, 2.25)
    ax.set_ylim(1e-3, 1e7)
    ax.set_yscale("log")

    ax.set_xlabel("Mass of Star")
    ax.set_ylabel("Orbital Period of Exoplanet")
    ax.set_title("Mass of Star vs. Orbital Period of Exoplanet (Per > {})".format(t0))

    filename = f"image_{i:03d}.png"
    path = "exoplanet_anim/" + filename
    fig.savefig(path)

    ax.cla()

fig.clf()
````
