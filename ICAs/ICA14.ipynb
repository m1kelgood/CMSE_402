```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data_files/human_development.csv")
data.tail(8)
```

```python
conts = data.iloc[189:,:]
```

```python
conts
```

```python
conts["Human Development Index (HDI)"]
```

```python
countries = conts["Country"].astype("category")
codes = countries.cat.codes

plt.scatter(conts["Gross National Income (GNI) per Capita"], conts["Mean Years of Education"], c=codes)

handles = []
for country, code in zip(countries.cat.categories, range(len(countries.cat.categories))):
    handles.append(plt.Line2D([], [], marker="o", linestyle="", color= plt.cm.tab20(code), label=country))

plt.legend(handles=handles, title="Country")
plt.xlabel("GNI per Capita")
plt.ylabel("Mean Years of Education")
plt.ylim(0, 18)
plt.savefig("GNI_vs_Ed.png")
plt.show()
```
