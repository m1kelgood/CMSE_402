```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data_files/congress-terms.csv")
```

```python
data
```

```python
data["age"].describe()
```

```python
fig, ax = plt.subplots(figsize=(6,6))
ax.hist(data["age"], bins=15, color="red", alpha=0.6)
plt.xlabel("Age", weight="bold")
plt.ylabel("Frequency", weight="bold")
plt.title("Senator Age Distribution", weight="bold")

ax.annotate('Mean Age (53)', xy=(51, 3100), xytext=(70, 2500),
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('Max Age (98)', xy=(96, 10), xytext=(75, 1000),
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('Min Age (25)', xy=(27, 30), xytext=(25, 1000),
            arrowprops=dict(facecolor='black', shrink=0.05))

fig.show()
```

```python
plt.hist2d(data["congress"], data["age"], cmap="terrain")
plt.xlabel("Congress Year")
plt.ylabel("Age of Individuals")
plt.show()
```
