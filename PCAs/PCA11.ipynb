````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import glob
````

## Part 1

````python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
````

````python
fig = plt.figure(figsize=(6,4))
for i in range(len(y)):
    ax = fig.add_subplot(1,1,1)
    ax.plot(x[:i], y[:i])
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1, 1)
    filename = "image_{:03d}.png".format(i)
    path = "sin_evolution/" + filename
    fig.savefig(path)
    fig.clf()
````

### Terminal Command Used
```text
ffmpeg -framerate 30 -i frames/image_%03d.png -pix_fmt video.mpg
```

````python
fig, ax = plt.subplots()
line, = ax.plot(x, np.sin(x))

def animate(i):
    line.set_ydata(np.sin(x + i / 50))
    return line,

ani = animation.FuncAnimation(fig, animate, interval=20, blit=True, save_count=50)
ani.save("sin_evo_MPL.mp4")
fig.clf()
````

````python
fig, ax = plt.subplots()
line, = ax.plot(x, np.sin(x))

def animate(i):
    line.set_ydata(np.sin(x + i / 50))
    return line,

ani = animation.FuncAnimation(fig, animate, interval=20, blit=True, save_count=50)
writer = animation.FFMpegWriter(fps=15, metadata=dict(artist="Me"),
                                bitrate=1800)
ani.save("sin_evo_FFMpeg.mp4", writer=writer)
fig.clf()
````
