```python
asc = pd.read_csv("acs2015_county_data.csv")
```

```python
mi = asc[asc["State"] == "Michigan"]
```

```python
mi
```

```python
categories = ["Male", "Female"]
values = [sum(mi["Men"]), sum(mi["Women"])]
plot = figure(height=400, width=600, x_range=categories, title="Michigan Population Breakdown by Gender")
plot.vbar(
    x=categories,
    top=values,
    width=0.6
    )


show(plot)
```

```python
commute = mi.iloc[:,24:30]
```

```python
commute
```

```python
commute.columns
```

```python
values = []
for i in list(commute.columns):
    values.append(sum(commute[i]))
```

```python
plot2 = figure(height=400, width=600, x_range=list(commute.columns), title="Michigan Work Breakdown by Commute Style")
plot2.vbar(
    x=list(commute.columns),
    top=values,
    width=0.6,
    color=(100, 200, 100)
    )
plot2.title_location = "above"

show(plot2)
```
