## Day 12 PCA
In this PCA, we'll be looking at ways to visualize distributions of data, via histograms, box plots, and violin plots.

```python
#our favorite imports
import matplotlib.pyplot as plt
import numpy as np
```

```python
#let's load in the data
A,B,C = np.loadtxt('data_file.csv', delimiter=',', skiprows=1, unpack=True)
```

### &#9989;&nbsp; **Question 1**
In the cell below, make **histograms** of our data A, B, and C from the data file. Choose a large number of bins so we can see the differences in the distributions. You're welcome to use <code>plt.hist()</code>. Be sure to add labels, so we can identify the different datasets. What do you notice about the data distributions?

```python
# Your code here!
plt.hist(A, bins=20)
plt.title("A")
plt.show()
```

```python
# Your code here!
plt.hist(B, bins=20)
plt.title("B")
plt.show()
```

```python
# Your code here!
plt.hist(C, bins=20)
plt.title("C")
plt.show()
```

A, B, and C are all distributed differently, as far as where there mean lies, the shape of their distribution, and how many modes they have.

### &#9989;&nbsp; **Question 2**
Now, let's make **box and whisker plots** of the same data. Plot them side by side, not overlapping. To accomplish this, recall that you can pass <code>plt.boxplot()</code> a list of data sets.  Be sure to add labels to the x-axis, so we can identify the different datasets.

```python
# Your code here!
plt.boxplot(x=[A,B,C])
plt.xticks([1,2,3], ["A","B","C"])
plt.show()
```

### &#9989;&nbsp; **Question 3**
Finally, let's make **violin plots** of the same data. This may be the first time you've plotted a violin plot, so here's the [documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.violinplot.html) for <code>plt.violinplot()</code>. We still want to plot the datasets side by side, not overlapping. Like with the box and whisker plots, you can pass <code>plt.violinplot()</code> a list of data sets. Again, be sure to add labels to the x-axis, so we can identify the different datasets. Unlike <code>plt.boxplot()</code>, <code>plt.violinplot()</code> does not have a <code>labels</code> parameter, so you'll need to set the x-axis labels manually. You can do this with the following code:

<code>plt.xticks([1, 2, 3], ['A', 'B', 'C'])</code>


```python
# Your code here!
plt.violinplot([A,B,C])
plt.xticks([1,2,3], ["A","B","C"])
plt.show()
```

### &#9989;&nbsp; **Question 4**

What do you notice about the differences between the three types of plots? Which do you think is the most informative? Why? Can you give an example or two when you might choose one over the other?

All of the plots communicate information about the distribution of the data in their own unique manner. I think that the violin plot is the most informative because it shows you the distribution of the plot and how it changes as it approaches the min and max values. While you can't identify the mean of the data on the violin plot (at least the one that I created), as displayed in the chapter of the book that corresponded with this PCA (The Truthful Art: Data, Charts, and Maps for Communication, by Alberto Cairo Chapter 11, Uncertainty and Significance), it gives you a better idea of how likely you are to expect a certain value as you deviate from the mean and go closer to the upper and lower bounds of the data. Because of this, I would choose a violin plot if I wanted to know how likely it is to expect a certain value in the data, such as the height of an individual given the distribution of the population.
